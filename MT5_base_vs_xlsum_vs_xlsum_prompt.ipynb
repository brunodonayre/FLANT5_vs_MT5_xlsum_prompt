```python
# 1. Librerías
import pandas as pd
import re
import torch
from transformers import T5Tokenizer, T5ForConditionalGeneration

# 2. Cargar CSV y preparar el DataFrame
ruta = r"C:\Users\HP\OneDrive - Grupo Alicorp\Escritorio\uni\Curso PLN\Cursos\PLN - ExamenFinal\Inputs\noticias_politicas_procesadas.csv"
df = pd.read_csv(ruta)
df.columns = df.columns.str.strip()

# Verificación de columna 'resumen_limpio'
if 'resumen_limpio' not in df.columns:
    print("⚠️ 'resumen_limpio' no encontrado. Usando 'Contenido' como respaldo.")
    df['resumen_limpio'] = df['Contenido']

# Extraer fecha desde la columna 'Contenido'
def extraer_fecha(texto):
    match = re.search(r"(\d{2})(\d{2})(\d{4})", str(texto))
    if match:
        return pd.to_datetime(f"{match.group(3)}-{match.group(2)}-{match.group(1)}")
    return pd.NaT

df["fecha"] = df["Contenido"].apply(extraer_fecha)
df = df.dropna(subset=["fecha"])

# 3. Cargar modelo FLAN-T5
## modelo FLAN-T5 localmente, descargado desde Hugging Face
modelo = "google/flan-t5-base"
tokenizer = T5Tokenizer.from_pretrained(modelo)
model = T5ForConditionalGeneration.from_pretrained(modelo)

# 4. Función de resumen
def resumir_con_flan(texto, instruccion="Haz un resumen claro y breve en español del siguiente contenido:"):
    if not texto or not isinstance(texto, str):
        return "Resumen no disponible."
    try:
        input_text = f"{instruccion} {texto}"
        inputs = tokenizer(input_text, return_tensors="pt", max_length=512, truncation=True)

        with torch.no_grad():
            outputs = model.generate(
                **inputs,
                max_length=60,
                num_beams=4,
                early_stopping=True
            )

        resumen = tokenizer.decode(outputs[0], skip_special_tokens=True)
        resumen = resumen.replace(instruccion, "").strip()
        return resumen
    except Exception as e:
        return f"⚠️ Error generando resumen: {e}"


# 5. Funciones de consulta
def ultimas_noticias_por_diario(df, n=3):
    noticias = (
        df.sort_values("fecha", ascending=False)
          .groupby("Diario")
          .head(n)
          .sort_values(["Diario", "fecha"], ascending=[True, False])
    )
    return noticias[["Diario", "Título", "fecha", "Enlace", "resumen_limpio"]]

def buscar_noticias_por_tema(df, tema, n=5):
    filtro = df["resumen_limpio"].str.contains(tema.lower(), case=False, na=False)
    resultados = df[filtro].sort_values("fecha", ascending=False).head(n)
    return resultados[["Diario", "Título", "fecha", "Enlace", "resumen_limpio"]]

# 6. Chatbot
def responder_pregunta(pregunta, df):
    pregunta = pregunta.lower()
    
    if "nuevo" in pregunta or "últimas" in pregunta:
        noticias = ultimas_noticias_por_diario(df)
        respuesta = "📰 Últimas noticias políticas por diario:\n\n"
        for _, row in noticias.iterrows():
            resumen_flan = resumir_con_flan(row["resumen_limpio"])
            respuesta += (
                f"- ({row['Diario']}) {row['Título']} ({row['fecha'].date()})\n"
                f"  📝 {resumen_flan}\n"
                f"  🔗 {row['Enlace']}\n\n"
            )
        return respuesta
    
    elif "dicen" in pregunta or "sobre" in pregunta:
        tema = pregunta.split("sobre")[-1].strip()
        noticias = buscar_noticias_por_tema(df, tema)
        if noticias.empty:
            return f"No encontré noticias recientes sobre “{tema}”."
        respuesta = f"🗞 Noticias recientes sobre “{tema}”:\n\n"
        for _, row in noticias.iterrows():
            resumen_flan = resumir_con_flan(row["resumen_limpio"])
            respuesta += (
                f"- ({row['Diario']}) {row['Título']} ({row['fecha'].date()})\n"
                f"  📝 {resumen_flan}\n"
                f"  🔗 {row['Enlace']}\n\n"
            )
        return respuesta
    
    else:
        return (
            "🤖 No entendí tu consulta. Puedes probar con:\n"
            "- ¿Qué hay de nuevo en política?\n"
            "- Dime las últimas noticias políticas\n"
            "- ¿Qué dicen los periódicos sobre Dina Boluarte?"
        )

# 7. Loop interactivo en consola
if __name__ == "__main__":
    print("🗣 Chatbot de noticias políticas con FLAN. Escribe 'salir' para terminar.\n")
    while True:
        user_input = input("Tú: ")
        if user_input.lower() in ["salir", "exit", "quit"]:
            print("👋 Hasta luego.")
            break
        respuesta = responder_pregunta(user_input, df)
        print("\n" + respuesta + "\n")
```

```python
import pandas as pd
import torch
from transformers import T5Tokenizer, T5ForConditionalGeneration
import evaluate

# === 1. Cargar benchmark con resumen_gold ===
import pandas as pd

# Ruta del archivo CSV con resumen_gold
ruta = r"C:\Users\HP\OneDrive - Grupo Alicorp\Escritorio\uni\Curso PLN\Cursos\PLN - ExamenFinal\Outputs\noticias_benchmark_resumen.csv"

df = pd.read_csv(ruta, sep=";", engine="python", encoding="utf-8-sig")
df = df.dropna(subset=["resumen_limpio", "resumen_gold"])




# === 2. Definir modelos y prompts ===
modelos = {
    "mt5_base": {
        "ruta": "google/mt5-base",
        "prompt": "Resume en español:"
    },
    "mt5_xlsum": {
        "ruta": "csebuetnlp/mT5_multilingual_XLSum",
        "prompt": "resumir:"
    }
}

# === 3. Función para resumir con un modelo dado ===
def generar_resumen(modelo_id, tokenizer, model, texto, prompt):
    entrada = f"{prompt} {texto}"
    inputs = tokenizer(entrada, return_tensors="pt", truncation=True, max_length=512)
    with torch.no_grad():
        output = model.generate(inputs["input_ids"], max_length=60, num_beams=4)
    return tokenizer.decode(output[0], skip_special_tokens=True)

# === 4. Generar resúmenes y evaluar con ROUGE ===
rouge = evaluate.load("rouge")
resultados = {}

for nombre, config in modelos.items():
    print(f"🧠 Procesando modelo: {nombre}")
    tokenizer = T5Tokenizer.from_pretrained(config["ruta"])
    model = T5ForConditionalGeneration.from_pretrained(config["ruta"])

    df[f"resumen_{nombre}"] = df["resumen_limpio"].apply(
        lambda t: generar_resumen(nombre, tokenizer, model, t, config["prompt"])
    )

    # Calcular métricas ROUGE
    scores = rouge.compute(
        predictions=df[f"resumen_{nombre}"].tolist(),
        references=df["resumen_gold"].tolist()
    )
    resultados[nombre] = scores

# === 5. Mostrar resultados ===
for nombre, metricas in resultados.items():
    print(f"\n📊 ROUGE para {nombre}:")
    for k, v in metricas.items():
        print(f"{k}: {v:.4f}")
```

##BERTScore F1 mide la similaridad semántica entre el resumen generado y el de referencia.
l valor va de 0 a 1:
> 0.85: Excelente (usualmente solo alcanzado por humanos o modelos finamente ajustados).
0.75 - 0.85: Muy bueno.
0.65 - 0.75: Aceptable, especialmente si los textos son difíciles o hay variación en redacción.
< 0.65: Puede haber problemas de cobertura o comprensión semántica.

```python
import evaluate

# Cargar métrica BERTScore
bertscore = evaluate.load("bertscore")

# Evaluar mt5_base
bert_mt5_base = bertscore.compute(predictions=df["resumen_mt5_base"].tolist(),
                                  references=df["resumen_gold"].tolist(),
                                  lang="es")

# Evaluar mt5_xlsum
bert_mt5_xlsum = bertscore.compute(predictions=df["resumen_mt5_xlsum"].tolist(),
                                   references=df["resumen_gold"].tolist(),
                                   lang="es")

# Mostrar resultados promedio
print("📊 BERTScore para mt5_base:")
print(f"Precision: {sum(bert_mt5_base['precision'])/len(bert_mt5_base['precision']):.4f}")
print(f"Recall:    {sum(bert_mt5_base['recall'])/len(bert_mt5_base['recall']):.4f}")
print(f"F1-score:  {sum(bert_mt5_base['f1'])/len(bert_mt5_base['f1']):.4f}")

print("\n📊 BERTScore para mt5_xlsum:")
print(f"Precision: {sum(bert_mt5_xlsum['precision'])/len(bert_mt5_xlsum['precision']):.4f}")
print(f"Recall:    {sum(bert_mt5_xlsum['recall'])/len(bert_mt5_xlsum['recall']):.4f}")
print(f"F1-score:  {sum(bert_mt5_xlsum['f1'])/len(bert_mt5_xlsum['f1']):.4f}")

#Apuntes:
#lang="es" indica que BERTScore usará un modelo multilingüe adecuado para español.
#La métrica más representativa es F1, que combina precisión y recall semántico.
#Esto te permitirá ver si aunque un modelo no use exactamente las mismas palabras (como FLAN), puede tener buena similaridad semántica.
```

```python
import pandas as pd
import torch
from transformers import T5Tokenizer, T5ForConditionalGeneration
import evaluate

# === 1. Cargar benchmark ===
ruta = r"C:\Users\HP\OneDrive - Grupo Alicorp\Escritorio\uni\Curso PLN\Cursos\PLN - ExamenFinal\Outputs\noticias_benchmark_resumen.csv"
df = pd.read_csv(ruta, sep=";", engine="python", encoding="utf-8-sig")
df = df.dropna(subset=["resumen_limpio", "resumen_gold"])

# === 2. Definir modelos y prompts ===
modelos = {
    "mt5_base": {
        "ruta": "google/mt5-base",
        "prompt": "Resume en español:"
    },
    "mt5_xlsum": {
        "ruta": "csebuetnlp/mT5_multilingual_XLSum",
        "prompt": "resumir:"
    },
    "mt5_xlsum_prompt": {
        "ruta": "csebuetnlp/mT5_multilingual_XLSum",
        "prompt": "Por favor, resume en español de forma clara y concisa el contenido de esta noticia:"
    }
}

# === 3. Función de resumen ===
def generar_resumen(modelo_id, tokenizer, model, texto, prompt):
    entrada = f"{prompt} {texto}"
    inputs = tokenizer(entrada, return_tensors="pt", truncation=True, max_length=512)
    with torch.no_grad():
        output = model.generate(inputs["input_ids"], max_length=60, num_beams=4)
    return tokenizer.decode(output[0], skip_special_tokens=True)

# === 4. Métricas ===
rouge = evaluate.load("rouge")
bertscore = evaluate.load("bertscore")
resultados = {}

# === 5. Evaluar modelos ===
for nombre, config in modelos.items():
    print(f"\n🧠 Procesando modelo: {nombre}")
    tokenizer = T5Tokenizer.from_pretrained(config["ruta"])
    model = T5ForConditionalGeneration.from_pretrained(config["ruta"])

    df[f"resumen_{nombre}"] = df["resumen_limpio"].apply(
        lambda t: generar_resumen(nombre, tokenizer, model, t, config["prompt"])
    )

    # ROUGE
    rouge_scores = rouge.compute(
        predictions=df[f"resumen_{nombre}"].tolist(),
        references=df["resumen_gold"].tolist()
    )

    # BERTScore
    bert_scores = bertscore.compute(
        predictions=df[f"resumen_{nombre}"].tolist(),
        references=df["resumen_gold"].tolist(),
        lang="es"
    )

    resultados[nombre] = {
        "ROUGE": rouge_scores,
        "BERTScore": {
            "precision": sum(bert_scores["precision"]) / len(bert_scores["precision"]),
            "recall": sum(bert_scores["recall"]) / len(bert_scores["recall"]),
            "f1": sum(bert_scores["f1"]) / len(bert_scores["f1"])
        }
    }

# === 6. Mostrar resultados ===
for nombre, res in resultados.items():
    print(f"\n📊 Resultados para {nombre}:")
    print("🔹 ROUGE:")
    for k, v in res["ROUGE"].items():
        print(f"  {k}: {v:.4f}")
    print("🔹 BERTScore:")
    for k, v in res["BERTScore"].items():
        print(f"  {k}: {v:.4f}")
```

### Interpretacion
1. El modelo mt5_xlsum sin modificar sigue siendo el mejor en ROUGE, sin embargo;
2. Con el prompt mejorado (mt5_xlsum_prompt) igualas (o mejoras muy ligeramente) en BERTScore, que capta mejor la calidad semántica.

| Modelo             | ROUGE-Lsum | BERTScore F1 | Comentario                                                     |
| ------------------ | ---------- | ------------ | -------------------------------------------------------------- |
| `mt5_base`         | 0.0495     | 0.5655       | Bajo rendimiento general. No entrenado para resumen.           |
| `mt5_xlsum`        | **0.1745** | 0.6589       | El mejor en ROUGE (más cercanía estructural).                  |
| `mt5_xlsum_prompt` | 0.1667     | **0.6590**   | Ligeramente mejor en BERTScore, gracias al prompt más natural. |


### En resumen:

- **ROUGE** mide solapamiento de *n-gramas*, útil para evaluar **fidelidad literal** al texto de referencia.
- **BERTScore** usa *embeddings* y es más tolerante a **reescrituras semánticas**, premiando la **calidad del lenguaje**.

**Entonces:**

> Si se buscan resúmenes más **legibles y naturales**, el prompt mejorado (`mt5_xlsum_prompt`) es una **gran elección**.
